## 1 LG ÏóêÎÑàÏßÄ ÏÜîÎ£®ÏÖò

#### üõú LG ÏóêÎÑàÏßÄ ÏÜîÎ£®ÏÖò Ï£ºÍ∞Ä Îç∞Ïù¥ÌÑ∞

```python
lg = pd.read_csv("LG·Ñã·Ö¶·Ü´·Ñâ·Ö©·ÜØ ·Ñå·ÖÆ·ÑÄ·Ö° ·ÑÉ·Ö¶·Ñã·Öµ·Ñê·Ö•.csv")
lg['Í∏∞ÏóÖÎ™Ö'] = 'LG ÏóêÎÑàÏßÄ ÏÜîÎ£®ÏÖò'  # Í∏∞ÏóÖÎ™Ö Ï∂îÍ∞Ä
lg.head()
```

```python
lg.info()

## ÌÉÄÏûÖ Î≥ÄÌôò
# [ÎÇ†Ïßú] object ‚Üí date time
# [Ï¢ÖÍ∞Ä, ÏãúÍ∞Ä, Í≥†Í∞Ä, Ï†ÄÍ∞Ä] object ‚Üí float
# [Í±∞ÎûòÎüâ] object ‚Üí int
# [Î≥ÄÎèô %] object ‚Üí float
```

ÎÇ†Ïßú Î≥ÄÌôò

```python
lg['ÎÇ†Ïßú'] = pd.to_datetime(lg['ÎÇ†Ïßú'])
```

Í±∞ÎûòÎüâ Ï†ïÏàò Î≥ÄÌôò

```python
# 1. Í±∞ÎûòÎüâÏóêÏÑú Ïñ¥Îñ§ Î¨∏ÏûêÍ∞Ä ÏûàÎäîÏßÄ Ï∂îÏ∂ú
lg['Í±∞ÎûòÎüâ'].str.extract('([A-Za-z]+)').value_counts()
```

```python
# 2. Î¨∏Ïûê 'K'Î•º ÏóÜÏï†Í≥†, Ïã§Ïàò Î≥ÄÌôò ÌõÑ, 1000 Í≥±Ìï®
lg['Í±∞ÎûòÎüâ'] = lg['Í±∞ÎûòÎüâ'].str.replace('K', '').apply(lambda x: float(x)) * 1000
```

#### üõú LG ÏóêÎÑàÏßÄ ÏÜîÎ£®ÏÖò ÌôòÏú® Î≥ÄÌôò(Ïõê ‚Üí Îã¨Îü¨)


```python
# ÎØ∏Íµ≠ ÌôòÏú® Îç∞Ïù¥ÌÑ∞ Îã§Ïö¥
kr_us = pd.read_csv("USD_KRW.csv")
kr_us = kr_us[['ÎÇ†Ïßú', 'Ï¢ÖÍ∞Ä']]
kr_us = kr_us.rename(columns = {"Ï¢ÖÍ∞Ä" : 'ÌôòÏú®Ï¢ÖÍ∞Ä'})
kr_us.head()
```

```python
kr_us.info()
## ÌÉÄÏûÖ Î≥ÄÌôò
# [ÎÇ†Ïßú] object ‚Üí date time
# [ÌôòÏú®Ï¢ÖÍ∞Ä] object ‚Üí float
```

```python
kr_us['ÎÇ†Ïßú'] = pd.to_datetime(kr_us['ÎÇ†Ïßú'])
kr_us.head()
```

```python
lg = pd.merge(lg, kr_us, how='left', left_on='ÎÇ†Ïßú', right_on='ÎÇ†Ïßú')
lg.head()
```

```python
lg.info()
```

```python
# Ï£ºÍ∞Ä Í¥ÄÎ†® Î≥ÄÏàò Î™®Îëê Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ object to float
def to_float(price):
    price = price.replace(',', '')
    price = float(price)
    return price

lg['Ï¢ÖÍ∞Ä'] = lg['Ï¢ÖÍ∞Ä'].apply(to_float)
lg['ÏãúÍ∞Ä'] = lg['ÏãúÍ∞Ä'].apply(to_float)
lg['Í≥†Í∞Ä'] = lg['Í≥†Í∞Ä'].apply(to_float)
lg['Ï†ÄÍ∞Ä'] = lg['Ï†ÄÍ∞Ä'].apply(to_float)
lg['ÌôòÏú®Ï¢ÖÍ∞Ä'] = lg['ÌôòÏú®Ï¢ÖÍ∞Ä'].apply(to_float)
lg.info()
```

```python
lg['Ï¢ÖÍ∞Ä'] = lg['Ï¢ÖÍ∞Ä'] / lg['ÌôòÏú®Ï¢ÖÍ∞Ä']
lg['ÏãúÍ∞Ä'] = lg['ÏãúÍ∞Ä'] / lg['ÌôòÏú®Ï¢ÖÍ∞Ä']
lg['Í≥†Í∞Ä'] = lg['Í≥†Í∞Ä'] / lg['ÌôòÏú®Ï¢ÖÍ∞Ä']
lg['Ï†ÄÍ∞Ä'] = lg['Ï†ÄÍ∞Ä'] / lg['ÌôòÏú®Ï¢ÖÍ∞Ä']
```

```python
lg.drop('ÌôòÏú®Ï¢ÖÍ∞Ä', axis=1, inplace=True)
```

```python
lg.head()
```

```python
lg.info()
```
