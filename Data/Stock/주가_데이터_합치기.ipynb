```python
import pandas as pd
```

## 1 LG 에너지 솔루션

#### 🛜 LG 에너지 솔루션 주가 데이터

```python
lg = pd.read_csv("LG엔솔 주가 데이터.csv")
lg['기업명'] = 'LG 에너지 솔루션'  # 기업명 추가
lg.head()
```

```python
lg.info()

## 타입 변환
# [날짜] object → date time
# [종가, 시가, 고가, 저가] object → float
# [거래량] object → int
# [변동 %] object → float
```

날짜 변환

```python
lg['날짜'] = pd.to_datetime(lg['날짜'])
```

거래량 정수 변환

```python
# 1. 거래량에서 어떤 문자가 있는지 추출
lg['거래량'].str.extract('([A-Za-z]+)').value_counts()
```

```python
# 2. 문자 'K'를 없애고, 실수 변환 후, 1000 곱함
lg['거래량'] = lg['거래량'].str.replace('K', '').apply(lambda x: float(x)) * 1000
```

#### 🛜 LG 에너지 솔루션 환율 변환(원 → 달러)


```python
# 미국 환율 데이터 다운
kr_us = pd.read_csv("USD_KRW.csv")
kr_us = kr_us[['날짜', '종가']]
kr_us = kr_us.rename(columns = {"종가" : '환율종가'})
kr_us.head()
```

```python
kr_us.info()
## 타입 변환
# [날짜] object → date time
# [환율종가] object → float
```

```python
kr_us['날짜'] = pd.to_datetime(kr_us['날짜'])
kr_us.head()
```

```python
lg = pd.merge(lg, kr_us, how='left', left_on='날짜', right_on='날짜')
lg.head()
```

```python
lg.info()
```

```python
# 주가 관련 변수 모두 데이터 타입 object to float
def to_float(price):
    price = price.replace(',', '')
    price = float(price)
    return price

lg['종가'] = lg['종가'].apply(to_float)
lg['시가'] = lg['시가'].apply(to_float)
lg['고가'] = lg['고가'].apply(to_float)
lg['저가'] = lg['저가'].apply(to_float)
lg['환율종가'] = lg['환율종가'].apply(to_float)
lg.info()
```

```python
lg['종가'] = lg['종가'] / lg['환율종가']
lg['시가'] = lg['시가'] / lg['환율종가']
lg['고가'] = lg['고가'] / lg['환율종가']
lg['저가'] = lg['저가'] / lg['환율종가']
```

```python
lg.drop('환율종가', axis=1, inplace=True)
```

```python
lg.head()
```

```python
lg.info()
```
